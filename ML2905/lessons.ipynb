### Task 1 (sns.regplot) : Plot the data and linear regression model fit into a scatterplot


sns.regplot is used to show both the scatter plot and linear regression model 

```python
sns.regplot(data = df, x = "sepal_length", y = "sepal_width")

plt.title("Regina: Sepal length vs Sepal width")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.show()
```

```python
sns.lmplot(data = df, x = "sepal_length", y = "sepal_width", hue = "species"
           ).set(title = "Regina: Sepal length vs Sepal width",
                 xlabel = "Sepal length",
                 ylabel = "Sepal width")

plt.show()
```

```python
sns.regplot(data = df, x = "petal_length", y = "petal_width")

plt.title("Regina: Petal length vs Petal width")
plt.xlabel("Petal length")
plt.ylabel("Petal width")

plt.show()
```

```python
sns.lmplot(data = df, x = "petal_length", y = "petal_width", hue = "species"
           ).set(title = "Regina: Petal length vs Petal width",
                 xlabel = "Petal length",
                 ylabel = "Petal width").set_style(rc = {'axes.facecolor': 'lightblue'})

plt.show()
```

```python
sns.set_style('dark')

sns.lmplot(data = df, x = "petal_length", y = "petal_width", hue = "species"
        ).set(title = "Regina: Petal length vs Petal width",
                xlabel = "Petal length",
                ylabel = "Petal width")

plt.show()
```

### Task 2 : GDP Scatterplot

```python
df = pd.read_csv('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
sns.set_style('whitegrid')

sns.scatterplot(data = df.loc[df.year == 1997], x = "gdpPercap", y = "lifeExp", size = 'pop', 
                sizes = (20,2000),
                hue = "country",
                legend = False
        ).set(title = "Regina: GDP per Capita vs Life expectancy, 1997",
                xlabel = "GDP per Capita",
                ylabel = "Life expectancy")

plt.show()
```

```python
sns.scatterplot(data = df.loc[(df.year == 2007) & (df.continent == "Asia")], 
                x = "gdpPercap", y = "lifeExp", size = 'pop', 
                sizes = (20,2000),
                hue = "country",
                palette = "crest",
                legend = False
        ).set(title = "Regina: GDP per Capita vs Life expectancy in Asia, 2007",
                xlabel = "GDP per Capita",
                ylabel = "Life expectancy")

plt.show()
```
